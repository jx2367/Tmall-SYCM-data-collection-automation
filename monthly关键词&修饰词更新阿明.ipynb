```python
import pandas as pd
import numpy as np
import os
import getpass
import sys
```

```python
account=getpass.getuser()
raw_path='/Users/'+str(account)+'/Desktop/weekly_kw_update/week_20210504-0510/'
path='/Users/'+str(account)+'/Desktop/weekly_kw_update/'
```

```python
cat_dict0=pd.read_excel(path+'dictionary.xlsx',sheet_name='cat_dict')
brand_dict0=pd.read_excel(path+'dictionary.xlsx',sheet_name='brand_dict')
```

```python
for file in os.listdir(raw_path):
    buff=file.split(sep="_")
    print(buff)
    break
```

```python
cat_dict=dict(zip(cat_dict0["subcat"],cat_dict0["cat"]))
brand_dict=dict(zip(brand_dict0["sub_kw"],brand_dict0["brand"]))
```

```python
cat_kw=dict()
cat_decrp=dict()
brand_kw=dict()
brand_descrp=dict()
for file in os.listdir(raw_path):
    buff=file.split(sep="_")
    if buff[2] in ['2021-05-03','2021-05-04']:
        df=pd.read_csv(raw_path+file)
    
        if buff[1].lower() in [x.lower() for x in cat_dict.keys()]:
            if "相关搜索" in buff[5]:
                print(buff[1])
                cat_kw[buff[1]]=df
            if "关联修饰" in buff[5]:
                cat_decrp[buff[1]]=df
        if buff[1].lower() in [x.lower() for x in brand_dict.keys()]:
            if "相关搜索" in buff[5]:
                brand_kw[buff[1]]=df
            if "关联修饰" in buff[5]:
                brand_descrp[buff[1]]=df
            
```

```python
cat_kw2=pd.concat(cat_kw,axis=0).reset_index()
cat_decrp2=pd.concat(cat_decrp,axis=0).reset_index()
brand_kw2=pd.concat(brand_kw,axis=0).reset_index()
brand_descrp2=pd.concat(brand_descrp,axis=0).reset_index()
```

```python
def add_w(data,year,month):
    data=data.loc[data["日期"].isna()==False]
    data["Year"]=year
    data["Week_Detail"]=data["日期"]
    data2=data[["Year","Week_Detail"]+[x for x in data.columns if x not in ["level_1","日期","Year","Week","Week_Detail"]]]
    return(data2)
```

```python
def add_yw(data,year,month):
    data=data.loc[data["日期"].isna()==False]
    data["Year"]=year
    data["Month"]=month
    data["Month_Detail"]=data["日期"]
    data2=data[["Year","Month","Month_Detail"]+[x for x in data.columns if x not in ["level_1","日期","Year","Week","Week_Detail"]]]
    return(data2)
```

```python
cat_kw3=add_yw(cat_kw2,2021,4)
cat_kw3=cat_kw3.rename(columns={"level_0":"keywords"},inplace=False)
cat_kw3["category"]=[cat_dict[x.lower()] for x in cat_kw3["keywords"]]
print(cat_kw3.head(3))


cat_decrp3=add_yw(cat_decrp2,2021,4)
cat_decrp3=cat_decrp3.rename(columns={"level_0":"keywords"},inplace=False)
cat_decrp3["category"]=[cat_dict[x.lower()] for x in cat_decrp3["keywords"]]
print(cat_decrp3.head(3))

brand_kw3=add_yw(brand_kw2,2021,4)
brand_kw3=brand_kw3.rename(columns={"level_0":"Search_keywords"},inplace=False)
brand_kw3["Brand"]=[brand_dict[x.lower()] for x in brand_kw3["Search_keywords"]]
print(brand_kw3.head(3))

brand_descrp3=add_yw(brand_descrp2,2021,4)
brand_descrp3=brand_descrp3.rename(columns={"level_0":"Search_keywords"},inplace=False)
brand_descrp3["Brand"]=[brand_dict[x.lower()] for x in brand_descrp3["Search_keywords"]]
print(brand_descrp3.head(3))
```

```python
# 获取ExcelWriter对象
pd.DataFrame().to_excel(path+"周度关键词&修饰词_阿明.xlsx")
writer = pd.ExcelWriter(path+"周度关键词&修饰词_阿明.xlsx",engine='xlsxwriter',options={'strings_to_numbers': True}) 
 
#将df1与df2写入writer中
brand_kw3[["Year","Month","Month_Detail","Brand"]+[x for x in brand_kw3.columns if x not in ["Year","Month","Month_Detail","Brand"]]].to_excel(writer, sheet_name='品牌关键词',index=False)
brand_descrp3[["Year","Month","Month_Detail","Brand"]+[x for x in brand_descrp3.columns if x not in ["Year","Month","Month_Detail","Brand"]]].to_excel(writer, sheet_name='品牌修饰词',index=False)
cat_kw3[["Year","Month","Month_Detail","category"]+[x for x in cat_kw3.columns if x not in ["Year","Month","Month_Detail","category"]]].to_excel(writer, sheet_name='Category关键词',index=False)
cat_decrp3[["Year","Month","Month_Detail","category"]+[x for x in cat_decrp3.columns if x not in ["Year","Month","Month_Detail","category"]]].to_excel(writer, sheet_name='Category修饰词',index=False)
 
#保存writer
writer.save() 
writer.close()
```
